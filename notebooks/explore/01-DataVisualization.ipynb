# Data Visualization

## Preamble

```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

## Load Data

```python
root_dir = Path().resolve().parents[1]
print(root_dir)
```

```python
with h5py.File(root_dir / 'data' / 'processed' / 'X.hdf5') as X_file:
    sinusoids = X_file['sinusoids'][:]

with h5py.File(root_dir / 'data' / 'processed' / 'y.hdf5') as y_file:
    frequencies = y_file['frequencies'][:].ravel()
    phases = y_file['phases'][:].ravel()
    amplitudes = y_file['amplitudes'][:].ravel()
```

## Visualize N=4 Samples

```python
N = 4

t = np.arange(100)/16000

plt.figure(figsize=(16,16))

for i in range(1,N+1):
    plt.subplot(4,1,i)
    plt.plot(t,sinusoids[i])
    plt.xlabel('time', fontsize=20)
    plt.ylabel('amplitude', fontsize=20)
    plt.title('f: {:.2f} Hz, $\phi$: {:.2f} rad, A: {:.2f}'.format(frequencies[i], phases[i], amplitudes[i]), fontsize=24)
    
plt.tight_layout()
plt.show()
```
